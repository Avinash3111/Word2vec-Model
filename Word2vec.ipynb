```python
import pandas as pd
data = [{'message':'i like the movie','sentiment':'positive'},
        {'message':'movie is not good','sentiment':'negative'},
        {'message':'wasted my time','sentiment':'negative'},
        {'message':'liked the movie','sentiment':'positive'}
        ]
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
messages = df['message']
labels = df['sentiment']
```

```python
import gensim.downloader as api
from gensim.models import Word2Vec
from sklearn.preprocessing import scale
import numpy as np
from sklearn.svm import SVC
```

```python
word2vec_model = api.load('glove-wiki-gigaword-100')
```

```python
t = " movie is very bad"
t.split()
```

```python
def feature_vector(text,model):
   words = text.split()
   feature_vec = np.zeros((model.vector_size,),dtype='float32')
   nwords =0
   for word in words:
     if word in model.key_to_index:
      nwords +=1
      feature_vec = np.add(feature_vec,model[word])

     if nwords > 0:
         feature_vec = np.divide(feature_vec,nwords)
     return feature_vec
```

```python
w2vec_features = np.array([feature_vector(text,word2vec_model) for text in messages])
```

```python
w2vec_features
```

```python
w2v_scaled_features = scale(w2vec_features)
```

```python
w2v_scaled_features
```

```python
from sklearn.svm import SVC
```

```python
clf = SVC(kernel='linear')
```

```python
clf.fit(w2v_scaled_features,labels)
```

```python
clf.score(w2v_scaled_features,labels)
```
